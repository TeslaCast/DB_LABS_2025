34.	Создайте функцию, рассчитывающую среднюю зарплату преподавателей в определенном структурном подразделении.

```python
CREATE OR REPLACE FUNCTION get_avg_salary_by_structural_unit(
    p_structural_unit_id INTEGER
)
RETURNS NUMERIC AS $$
BEGIN
    RETURN (
        SELECT AVG(p.salary::NUMERIC)::NUMERIC
        FROM public.professors p
        JOIN public.employments e ON p.professor_id = e.professor_id
        WHERE e.structural_unit_id = p_structural_unit_id
    );
END;
$$ LANGUAGE plpgsql;
```

```python
SELECT get_avg_salary_by_structural_unit(1);
```
