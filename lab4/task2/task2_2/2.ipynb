14.	Напишите скрипт, отбирающий случайным образом 10 студентов из списка студентов 3-го курса, имеющих больше 10 двоек для участия в субботнике. Выведите фамилию имя и количество двоек у студента.

```python
DO $$
DECLARE
    r RECORD;
BEGIN
    FOR r IN 
        SELECT last_name, first_name, COUNT(*) AS dvojki_count
        FROM public.students s
        JOIN public.field_comprehensions fc ON s.student_id = fc.student_id
        WHERE s.students_group_number LIKE '%-3_'
        AND fc.mark = 2 -- Двоек
        GROUP BY s.student_id, s.last_name, s.first_name
        HAVING COUNT(*) > 10
        ORDER BY RANDOM()
        LIMIT 10
    LOOP
        RAISE NOTICE 'Студент: %, %, двоек: %', r.last_name, r.first_name, r.dvojki_count;
    END LOOP;
END $$;
```
