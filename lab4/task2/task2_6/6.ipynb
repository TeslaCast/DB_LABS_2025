54.	Создайте триггер, который запрещает изменение в структурных подразделениях 

```python
CREATE OR REPLACE FUNCTION prevent_structural_unit_changes()
RETURNS TRIGGER AS $$
BEGIN
    RAISE EXCEPTION 'Изменение данных в таблице structural_units запрещено';
    RETURN NULL;
END;
$$ LANGUAGE plpgsql;

CREATE OR REPLACE TRIGGER trg_prevent_structural_unit_changes
BEFORE UPDATE OR DELETE ON public.structural_units
FOR EACH ROW EXECUTE FUNCTION prevent_structural_unit_changes();
```
